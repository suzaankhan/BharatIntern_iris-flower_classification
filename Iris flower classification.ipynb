## Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
import seaborn as sns
# from urllib.request import urlretrieve/

%matplotlib inline
```

# Assigning Column names

```python
columns = ['sepal_l', 'sepal_w', 'petal_l', 'petal_w', 'class']
iris_df.columns = columns
```

```python
iris_df.head()
```

# Data analysis and visualisation

```python
sns.pairplot(iris_df, hue=iris_df.columns.all())
plt.show()
```

```python
sns.heatmap(iris_df.corr(), annot=True)
```

```python
data = iris_df.values  #return numpy array
#var = data[y-axis, x-axis]
x = data[:, 0:4]
y = data[:, 4]
```

```python
x
```

# Splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=101)
```

```python
x_train
```

```python
y_train
```

# Training the classification model using Support vector algorithm

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(x_train, y_train)
```

# Prediction and Accuracy for SVC

```python
svc_prediction = svc_model.predict(x_test)
```

```python
svc_acc = accuracy_score(y_test, svc_prediction)
```

```python
print(f'Accuracy: {svc_acc}')
```

# Training the classification model using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

# Prediction and Accuracy for LR model

```python
prediction = model.predict(x_test)
```

```python
lr_acc = accuracy_score(y_test, prediction)
```

```python
print(f'Accuracy: {lr_acc}')
```
